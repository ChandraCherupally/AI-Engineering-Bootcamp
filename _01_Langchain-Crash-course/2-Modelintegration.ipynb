## Model Integration With OpenAI, Gemini & GROQ

### OpenAI Model Integration
#### Method1

```python
import os
from dotenv import load_dotenv

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("gpt-4.1")
model
```

```python
response = model.invoke("Hello how are you?")
response.content
```

#### Method2

```python
from langchain_openai import ChatOpenAI 
model = ChatOpenAI(model="gpt-4.1")
response = model.invoke("What is the capital of India")
response.content
```

### Google gemini Model Integration

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
model = init_chat_model("google_genai:gemini-2.5-pro")
response = model.invoke("Capital of India")
response.content
```

#### Method2

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=1.0,  # Gemini 3.0+ defaults to 1.0
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)
response = model.invoke("Capital of India")
response.content
```

### Groq Model Integration

#### Method1

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Capital of India")
response.content
```

#### Method2

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=1.0,  # Gemini 3.0+ defaults to 1.0
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)
response = model.invoke("Capital of India")
response.content
```

## Streaming and Batch

### Streaming

```python
for chunk in model.stream("Write me a 200 words paragraph on AI"):
    print(chunk.text, end="|", flush=True)
```

```python
for chunk in model.stream("Why do parrots have colorful feathers"):
    print(chunk.text, end="|", flush=True)
```

### Batch

```python
responses = model.batch(
    ["What is the capital of India?",
    "Who won 1983 cricket wold cup?",
    "What is quantum computing explain in 20 words?",
    ],
    config={
        'max_concurrency':5, # Limit to 5 parallel calls
    }
)

for response in responses:
    print(response)
```
